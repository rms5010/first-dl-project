# 🎯 유튜브 트렌드 영상 분석 - 중간 보고서

---

## 1. 프로젝트 개요

- **주제**: 유튜브 트렌드 영상 분석  
- **목적**:  
  사람들이 빠르게 최신 트렌드를 파악할 수 있도록,  
  최근 업로드된 유튜브 영상 중 **조회수가 높은 콘텐츠**를 분석하여  
  **인기 키워드, 카테고리별 경향, 콘텐츠 특성**을 시각화하고 요약한다.

---

## 2. 중간 결과

### - 현재까지 완료된 작업

- 유튜브 트렌드 데이터를 활용해 EDA를 완료함.
- 카테고리별 평균 조회수, 좋아요 비율, 시간대별 업로드 패턴, 제목 키워드 등 다양한 변수에 대한 탐색적 분석 수행.
- 특히 영상별 조회수의 **시간에 따른 변화**를 기반으로 시계열 데이터를 구성하였고, **딥러닝 기반 예측 모델 적용을 위한 전처리**도 완료됨.

---

## 3. 수정 목표

- 기존 목표: 영상 정보(카테고리, 업로드 시간, 제목 길이, 태그 수 등)와 인기 지표(조회수, 좋아요 등)의 관계 분석, LSTM 모델을 활용한 시계열 기반 트렌드 예측, KMeans 알고리즘을 통한 영상 유형 클러스터링
- 수정 후 목표:
  - **시계열 예측 모델(LSTM)을 활용하여 미래 조회수 예측**
  - **영상의 특성 기반 클러스터링(KMeans)**을 통해 트렌드 유형 분류(가능한 경우에만)
- 예측을 위한 시계열 데이터는 `video_id`별 `view_count`를 시간순으로 정렬하여 구성
- 클러스터링은 EDA 결과를 바탕으로 추출된 변수(feature)를 정규화하여 적용할 예정

---

## 4. 추후 일정 (총 4주차 내 마감)

| 주차 | 작업 내용 |
|------|-----------|
| 1주차 | 시계열 데이터 구성 (`video_id`별 view_count 정렬), 시각적 확인 |
| 2주차 | LSTM 모델 설계 및 학습 데이터 생성, 학습 진행 |
| 3주차 | 예측 결과 시각화 및 모델 성능(MAE/MSE) 평가 |
| 4주차 | (선택) KMeans 클러스터링 분석 및 군집 해석, 발표자료 및 보고서 작성 |

---

## 5. 예상 결과 및 기대 효과

- **LSTM 기반 예측 모델 결과**
  - 시간 경과에 따른 조회수 변화 패턴을 반영한 예측이 가능함
  - 단순 회귀보다 시계열 정보를 반영함으로써 예측 정확도를 높일 수 있음
  - 예측 결과를 통해 영상 업로드 시점에 따른 성과를 사전 예측하거나, 트렌드 지속성을 파악할 수 있음

- **KMeans 클러스터링 결과 (선택 분석)**
  - 영상들의 feature(조회수, 좋아요 비율 등)를 바탕으로 유사한 유형별 그룹 도출 가능
  - 군집별 대표 특성을 분석하여, 조회수가 높은 유형이나 반응이 좋은 영상 스타일 파악 가능
  - 향후 콘텐츠 기획 시 어떤 요소가 인기를 끌 가능성이 높은지 전략적으로 참고 가능

- **종합적 기대 효과**
  - 유튜브 영상의 성공 요소(카테고리, 키워드, 업로드 시간 등)를 정량적으로 분석함으로써,
    **콘텐츠 기획 및 마케팅 전략에 인사이트 제공**
  - 단순 분석을 넘어 **예측과 분류까지 포함된 통합형 데이터 분석 프로젝트**로서 포트폴리오 활용도 높음
  - 세대와 관계없이 유행을 쉽게 파악하고, 함께 즐길 수 있도록 하겠다는 본 목적에 부합함.
